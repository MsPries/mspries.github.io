```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(1968080)

num = 5
contrast = []
for i in range(num):
    contrast.append(i*2)
# constrast = [(i*4+12) for i in range(num)] # -- shortened version
error = np.random.rand(num)
colors = 'blue'
area = np.pi * 8

plt.scatter(contrast, error, s=area, c=colors, alpha=0.5) # automatically adjusts axes based on numbers
plt.show()
```
